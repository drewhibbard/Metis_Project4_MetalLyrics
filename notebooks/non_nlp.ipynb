```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import pickle
```

```python
with open('backup/metal1.pickle','rb') as rf:
    metal = pickle.load(rf)

with open('cv_dataframe.pickle','rb') as rf:
    cv_df = pickle.load(rf)
```

```python
metal = pd.DataFrame(metal)
```

```python
metal = pd.concat([metal,cv_df],axis=1)
```

```python
metal.head()
```

```python
metal.dropna(subset=['release_date'],inplace=True)
```

```python
metal['year'] = metal.release_date.str[:4]
```

```python
metal = metal[metal.year!='0001']
metal = metal[metal.year!='0216']
```

```python
metal['decade'] = metal.year.str[:3] + "0's"
```

```python
metal.decade.value_counts()
```
